# One type of login project

```python
usernames=["Aflah","123","op"]
passwords=["111","msalhop","246"]
a=input("Enter Username:")
for i in range(1,6):
    if a in usernames:
        break;
    else:
        print("Invalid Username")
        break;
if a in usernames:
    b=input("Enter Password:")
    for i in range(1,6):
        if b in passwords:
            break;
        else:
            print("Invalid Password")
            break;
    for i in range(len(usernames)):
        if a==usernames[i] and b==passwords[i]:
            print("Login Succesfully")
            break;
        else:
            print("Invalid Login")
            break;
```
